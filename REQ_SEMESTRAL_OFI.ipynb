### CLEANING PHONES

```python
def normalizar(phone):
    #cambiar a cadena.
    if type(phone) is float:
        if not math.isnan(phone):
            phone =(str)((int)(phone)) 
        else:
            phone = "0"
    else:        
        #eliminar espacios
        phone = phone.replace(" ", "")
        #eliminar #
        phone = phone.replace("#", "")
        #eliminar .
        phone = phone.replace(".", "")
        #eliminar .
        phone = phone.replace("(", "")
        #eliminar .
        phone = phone.replace(")", "")
        #eliminar .
        phone = phone.split("-")[0]
        #eliminar +51
        phone = phone.replace("+51", "")
        #eliminar 051
        phone = phone[3:] if phone[0:3] == "051" else phone
        #eliminar 0051
        phone = phone[4:] if phone[0:4] == "0051" else phone
        #eliminar 0051
        phone = phone[2:] if phone[0:2] == "51" else phone  
        #eliminar 0 ADELANTE
        phone = phone[1:] if phone[0:1] == "0" else phone  
        #eliminar ,
        phone = phone.replace(",","")
    return phone
```

```python
def validaremail(email):
    nuevo_email = ""
    if type(email) is float:
        nuevo_email = (str)(email)
    else:
        if  email.find("@") > -1 and email.find(".") > -1:    
           nuevo_email = email
        else: 
           nuevo_email = ""    
    return nuevo_email
```
